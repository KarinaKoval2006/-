```python
import random

class Character:
    def __init__(self, name, health, damage):
        self.name = name
        self.health = health
        self.damage = damage

class Enemy(Character):
    def __init__(self, name):
        super().__init__(name, random.randint(30, 80), random.randint(5, 15))

class Player(Character):
    def __init__(self, name):
        super().__init__(name, 100, 10)
        self.coins = 0

    def heal(self):
        if self.coins >= 10:
            self.coins -= 10
            self.health += 40
            print(f"{self.name} использовал зелье лечения и восстановил 20 здоровья!")
            print(f"Текущее здоровье: {self.health}")
        else:
            print("Недостаточно монет для покупки зелья лечения!")

class Game:
    def __init__(self):
        self.player = self.create_player()
        self.locations = {
            '1': self.village,
            '2': self.field,
            '3': self.lake,
            '4': self.forest,
            '5': self.dungeon
        }
        self.enemies = [
            Enemy("Водяной"),
            Enemy("Оборотень"),
            Enemy("Зомби")
        ]
        self.game_over = False

    def create_player(self):
        name = input("Введите ваше имя: ")
        return Player(name)

    def village(self):
        print(f"Вы находитесь в деревне. Здесь нет врагов.")
        action = input("Хотите посетить лекаря (1) или магазин (2)? ")
        if action == '1':
            self.player.health = 100
            print(f"{self.player.name} вылечился до максимума!")
        elif action == '2':
            self.player.coins += 0  # Начальные монеты в магазине
            print(f"Вы сейчас имеете {self.player.coins} монет.")
            self.player.heal()

    def lake(self):
        self.battle(random.choice([self.enemies[0]]))

    def forest(self):
        self.battle(random.choice([self.enemies[1]]))

    def field(self):
        self.battle(random.choice([self.enemies[2]]))

    def dungeon(self):
        print("Вы вошли в подземелье.")
        dragon = Enemy("Огнедышащий дракон")
        dragon.health = 300
        self.battle(dragon)

    def battle(self, enemy):
        print(f"Встретили врага: {enemy.name} с {enemy.health} здоровья и {enemy.damage} урона!")
        while enemy.health > 0 and self.player.health > 0:
            choice = input("Выберите действие: (1) атака или (2) бегство: ")
            if choice == '1':
                enemy.health -= self.player.damage
                print(f"{self.player.name} атакует {enemy.name}! У врага осталось {enemy.health} здоровья.")
                if enemy.health > 0:
                    self.player.health -= enemy.damage
                    print(f"{enemy.name} атакует {self.player.name}! У вас осталось {self.player.health} здоровья.")


            elif choice == '2':
                print(f"{self.player.name} убегает от {enemy.name}!")
                return  # Бегство
            else:
                print("Некорректный выбор. Попробуйте снова.")

        if self.player.health <= 0:
              print("Вы погибли! Конец игры.")
              self.game_over = True
        else:
              self.player.coins += 5
              self.player.damage += 10
              self.player.health += 20
              if self.player.health > 200:
                self.player.health = 200  # Приведение здоровья к максимуму
                print(f"{enemy.name} побежден!")
                print(f"{self.player.name} получает 5 монет и увеличивает урон на 10.")

              if enemy.name == "Огнедышащий дракон":
                print("Вы победили огнедышащего дракона! Игра окончена.")
                self.game_over = True

              if self.player.health >= 200:
                print("Вы достигли 200 здоровья! Теперь вы можете войти в подземелье.")
                self.locations['5']()  # Автоматический переход в подземелье

    def start_game(self):
        while not self.game_over:

            print("\nВыберите локацию:")
            print("1 - Деревня")
            print("2 - Поле")
            print("3 - Озеро")
            print("4 - Лес")
            print("5 - Подземелье")
            location = input("Введите номер локации: ")
            if location in self.locations:
                self.locations[location]()
            else:
                print("Некорректный номер локации.")

if __name__ == "__main__":
    game = Game()
    game.start_game()
```

```python
#импорт библиотеки random
import random

#класс персонажа (для любого в будущем) (имя, здоровье, урон - атрибуты)
class Character:
    def __init__(self, name, health, damage):
        self.name = name
        self.health = health
        self.damage = damage

#класс для врагов на основе character
class Enemy(Character):
    def __init__(self, name):
        super().__init__(name, random.randint(30, 80), random.randint(5, 15))

#класс для игрока на основе character
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 100, 10)
        self.coins = 0
        self.max_health = 100 #потом пригодится для увеличения хп

    def heal(self): #метод для лечения игрока
        self.health = self.max_health
        print(f"{self.name} вылечился до максимума! Текущее здоровье: {self.health}")

#новый класс - игра
class Game:
    def __init__(self):
        self.player = self.create_player() #создание игрока (без скобок не будет работать метод)
        self.enemies = {
            "lake": Enemy("Водяной"),
            "forest": Enemy("Оборотень"),
            "field": Enemy("Зомби")
        } #{} - словарь
        self.locations = ["village", "field", "lake", "forest"] #список локаций
        self.game_over = False #флаг окончания игры (пока False - игра продолжается)

#ввыбор имени
    def create_player(self):
        name = input("Введите ваше имя: ")
        return Player(name) #возвращает нового игрока с введённым именем


    def village(self):
        print(f"Вы находитесь в деревне. Здесь нет врагов.")
        action = input("Что хотите сделать? (1) Посетить лекаря, (2) Проверить здоровье: ")
        if action == '1':
            self.player.heal() #восстановление хп
        elif action == '2':
            print(f"Текущее здоровье: {self.player.health}, Максимальное здоровье: {self.player.max_health}") #проверка характеристик
        else:
            print("Некорректный выбор. Попробуйте снова.")


    def battle(self, enemy):
        print(f"Встретили врага: {enemy.name}, здоровье: {enemy.health}, урон: {enemy.damage}")
        print(f"Ваше здоровье: {self.player.health}, ваш урон: {self.player.damage}")

        while enemy.health > 0 and self.player.health > 0: #продолжение боя, пока живы
            choice = input("Выберите действие: (1) атака, (2) бегство: ")
            if choice == '1':
                enemy.health -= self.player.damage
                print(f"{self.player.name} атакует {enemy.name}! У врага осталось {max(0, enemy.health)} здоровья.") #не даёт врагу быть с отрицательным хп
                if enemy.health > 0:
                    self.player.health -= enemy.damage
                    print(f"{enemy.name} атакует {self.player.name}! У вас осталось {max(0, self.player.health)} здоровья.")
            elif choice == '2':
                print(f"{self.player.name} убегает от {enemy.name}!")
                return #завершает выполнение метода battle()
            else:
                print("Некорректный выбор. Попробуйте снова.")

        if self.player.health <= 0: #проигрыш
            print("Вы погибли! Конец игры.")
            self.game_over = True
        else:
            print(f"Вы победили {enemy.name}!")
            self.player.coins += 5
            self.player.damage += 10
            self.player.max_health += 35  #увеличение максимального здоровья
            self.player.health = min(self.player.health + 20, self.player.max_health) #частичное восстановление хп
            print(f"+5 монет, +10 урона, +35 к максимальному здоровью. Новое максимальное здоровье: {self.player.max_health}")
            print(f"Текущее здоровье: {self.player.health}")

    #подземелье с драконом
    def dungeon(self):
        dragon = Enemy("Огнедышащий дракон")
        dragon.health = 300
        dragon.damage = 25
        print("Вы вошли в подземелье. Перед вами Огнедышащий дракон!")
        self.battle(dragon)
        if not self.game_over: #если игра идёт - выводятся слова
            print("Вы победили огнедышащего дракона! Вы спасли мир!")
            restart = input("Хотите начать заново? (да/нет): ").lower()
            if restart == "да":
                self.__init__()
            else:
                self.game_over = True


    def start_game(self):
        print(f"Добро пожаловать в игру, {self.player.name}!")
        while not self.game_over: #while - начало цикла, который будет работать, пока условие истинно
            if self.player.max_health >= 200 and "dungeon" not in self.locations: #подземелье ещё не добавлено в список локаций
                self.locations.append("dungeon") #условия выполнены, добавляет "dungeon" в список доступных локаций

            print("\nВыберите локацию:")
            for i, location in enumerate(self.locations, 1): #список (self.locations), второй аргумент 1 - нумерация с 1
                print(f"{i} - {location.capitalize()}") #подставляет номер пункта

            try: #чтобы игра не сломалась, если пользователь введёт что-то не то
                choice = int(input("Введите номер локации: ")) - 1 #вычитает 1, потому что списки нумеруются с 0, а пользователь видит меню с нумерацией с 1
                if 0 <= choice < len(self.locations): #чтобы не было отриц чисел
                    location = self.locations[choice]
                    if location == "village":
                        self.village()
                    elif location == "dungeon":
                        self.dungeon()
                    else:
                        enemy = self.enemies.get(location) #пытается получить врага из словаря по ключу-локации
                        if enemy and enemy.health > 0:
                            self.battle(enemy)
                        else:
                            print(f"В этой локации больше нет врагов.")
                else:
                    print("Некорректный выбор. Попробуйте снова.")
            except ValueError:
                print("Введите число!")


if __name__ == "__main__": #cоздается объект игры и запускается основной цикл
    game = Game()
    game.start_game()
```
